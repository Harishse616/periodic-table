```python
from itertools import combinations
metals = df[df['symbol'].isin([ "Li", "Be", "Na", "Mg", "Al", "K", "Ca", "Sc", "Ti", "V", "Cr", "Mn",
    "Fe", "Co", "Ni", "Cu", "Zn", "Ga", "Rb", "Sr", "Y", "Zr", "Nb", "Mo",
    "Tc", "Ru", "Rh", "Pd", "Ag", "Cd", "In", "Sn", "Cs", "Ba", "La", "Ce",
    "Pr", "Nd", "Pm", "Sm", "Eu", "Gd", "Tb", "Dy", "Ho", "Er", "Tm", "Yb",
    "Lu", "Hf", "Ta", "W", "Re", "Os", "Ir", "Pt", "Au", "Hg", "Tl", "Pb",
    "Bi", "Fr", "Ra", "Ac", "Th", "Pa", "U", "Np", "Pu", "Am", "Cm", "Bk",
    "Cf", "Es", "Fm", "Md", "No", "Lr"])]


non_metals = df[df['symbol'].isin([ "H", "C", "N", "O", "F", "P", "S", "Cl", "Se", "Br", "I", "At"])]
```

```python
import pandas as pd
from math import gcd

# Suppose df has: ['name', 'symbol', 'oxidation_states']

compounds = []

for _, elem1 in df.iterrows():
    ox1_list = [int(s.strip('+')) for s in elem1['oxidation_states'].strip("[]").split(',') if s.strip()]

    for _, elem2 in df.iterrows():
        ox2_list = [int(s.strip('+')) for s in elem2['oxidation_states'].strip("[]").split(',') if s.strip()]

        for charge1 in ox1_list:
            for charge2 in ox2_list:
                if charge1 * charge2 < 0:  # opposite charges attract
                    lcm = abs(charge1 * charge2) // gcd(charge1, charge2)
                    n1 = lcm // abs(charge1)
                    n2 = lcm // abs(charge2)
                    formula = f"{elem1['symbol']}{n1 if n1 > 1 else ''}{elem2['symbol']}{n2 if n2 > 1 else ''}"
                    compounds.append(formula)

                    # inside your loop…
                    if elem1['symbol'] == elem2['symbol']:
                        # same element case → H2, O2, N2, etc.
                        compound = f"{elem1['symbol']}2"
                    else:
                        # different elements → NaCl, Fe2O3, etc.
                        compound = (
                            f"{elem1['symbol']}{n1 if n1 > 1 else ''}"
                            f"{elem2['symbol']}{n2 if n2 > 1 else ''}"
                        )

                    compounds.append(compound)

# Handle diatomic molecules separately (H2, O2, Cl2, etc.)
for _, elem in df.iterrows():
    if elem['symbol'] in ['H','N','O','F','Cl','Br','I']:  # common diatomics
        compounds.append(f"{elem['symbol']}2")

for i in compounds:
  if i == 'HH' or i == 'NN' or i == 'OO' or i == 'FF' or i == 'ClCl' or i == 'BrBr' or i == 'II':
    compounds.remove(i)

for i in compounds:
  if i == i[::1]:
    compounds.remove(i)

print(compounds)
```
