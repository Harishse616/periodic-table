```python
symbol = input("enter the symbol of element : ")
def get_element_details(symbol):
  symbol = symbol.capitalize()
  row = df[df['symbol'] == symbol]
  if row.empty:
    return f"No element found with symbol '{symbol}'"
  else:
    return row.iloc[0].to_dict()

details = get_element_details(symbol)
print(details)
```

```python
from itertools import combinations
metals = df[df['symbol'].isin([ "Li", "Be", "Na", "Mg", "Al", "K", "Ca", "Sc", "Ti", "V", "Cr", "Mn",
    "Fe", "Co", "Ni", "Cu", "Zn", "Ga", "Rb", "Sr", "Y", "Zr", "Nb", "Mo",
    "Tc", "Ru", "Rh", "Pd", "Ag", "Cd", "In", "Sn", "Cs", "Ba", "La", "Ce",
    "Pr", "Nd", "Pm", "Sm", "Eu", "Gd", "Tb", "Dy", "Ho", "Er", "Tm", "Yb",
    "Lu", "Hf", "Ta", "W", "Re", "Os", "Ir", "Pt", "Au", "Hg", "Tl", "Pb",
    "Bi", "Fr", "Ra", "Ac", "Th", "Pa", "U", "Np", "Pu", "Am", "Cm", "Bk",
    "Cf", "Es", "Fm", "Md", "No", "Lr"])]


non_metals = df[df['symbol'].isin([ "H", "C", "N", "O", "F", "P", "S", "Cl", "Se", "Br", "I", "At"])]
```

```python
from itertools import combinations
from math import gcd # Import gcd function

def generate_binary_compouds(metals, non_metals):
  compounds = []

  for _, metal in metals.iterrows():
    # Process oxidation_states as a list
    metal_states = [int(s.strip('+')) for s in metal['oxidation_states'].strip("[]").split(',')]
    for _, non_metal in non_metals.iterrows():
      # Process oxidation_states as a list and fix the typo
      non_metal_states = [int(s.strip('+')) for s in non_metal['oxidation_states'].strip("[]").split(',')]
      for m_states in metal_states:
        for nm_state in non_metal_states:
          if m_states + nm_state == 0:
            compound = f"{metal['symbol']}{non_metal['symbol']}"
            compounds.append(compound)
          else :
            lcm = abs(m_states*nm_state)// gcd(m_states, nm_state)
            m_count = lcm // abs(m_states)
            nm_count = lcm // abs(nm_state)
            compound = f"{metal['symbol']}{m_count if m_count>1 else ''}{non_metal['symbol']}{nm_count if nm_count>1 else ''}"
            compounds.append(compound)

  return compounds

# Removed the redundant gcd function definition as it's imported from math
binary_compounds = generate_binary_compouds(metals, non_metals)
print(binary_compounds)
```
